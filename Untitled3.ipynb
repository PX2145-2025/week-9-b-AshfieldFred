```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def orbit_solve(r0, v0, tmax, N, method='euler'):
    import numpy as np

    dt = tmax / (N - 1)
    t = np.linspace(0, tmax, N)

    r = np.zeros((N, 2))
    v = np.zeros((N, 2))
    a = np.zeros((N, 2))

    r[0] = r0
    v[0] = v0

    G = 1
    M = 1

    # initial acceleration
    dist = np.linalg.norm(r[0])
    a[0] = -G * M * r[0] / dist**3

    for i in range(N - 1):

        if method == 'euler':
            r[i+1] = r[i] + v[i] * dt
            v[i+1] = v[i] + a[i] * dt

        elif method == 'euler-cromer':
            v[i+1] = v[i] + a[i] * dt
            r[i+1] = r[i] + v[i+1] * dt

        elif method == 'verlet':
            r[i+1] = r[i] + v[i] * dt + 0.5*a[i]*dt**2
            dist_next = np.linalg.norm(r[i+1])
            a[i+1] = -G * M * r[i+1] / dist_next**3
            v[i+1] = v[i] + 0.5 * (a[i] + a[i+1]) * dt

        dist_next = np.linalg.norm(r[i+1])
        a[i+1] = -G * M * r[i+1] / dist_next**3

    return t, r, v
```

```python
r0 = np.array([10.0, 0.0])
v0 = np.array([0.0, np.sqrt(1/10)])

t, r, v = orbit_solve(r0, v0, 400, 100,method='euler') #T=400 for 2 orbits
t1, r1, v1 = orbit_solve(r0, v0, 400, 200,method='euler')
t2, r2, v2 = orbit_solve(r0, v0, 400, 400,method='euler')
plt.plot(r[:,0], r[:,1], label='N=100')
plt.plot(r1[:,0], r1[:,1],label='N=200')
plt.plot(r2[:,0], r2[:,1],label='N=400')
plt.xlabel('X-position')
plt.ylabel('Y-position')
plt.title('2 orbits using different N')
plt.legend()
plt.grid()
plt.show()
```

```python
t3, r3, v3 = orbit_solve(r0, v0, 400, 200,method='euler')
t4, r4, v4 = orbit_solve(r0, v0, 400, 200,method='euler-cromer')
t5, r5, v5 = orbit_solve(r0, v0, 400, 200,method='verlet')
plt.plot(r3[:,0], r3[:,1], label='euler')
plt.plot(r4[:,0], r4[:,1],label='euler-cromer')
plt.plot(r5[:,0], r5[:,1],label='verlet')
plt.xlabel('X-position')
plt.ylabel('Y-position')
plt.title('Different methods of integration')
plt.legend()
plt.grid()
plt.show()
```
